# 1.2.1 - Preparing loop:

```python
import requests
from bs4 import BeautifulSoup

livros = []
url_base = 'https://www.amazon.com.br/gp/bestsellers/books/?ie=UTF8&ref_=sv_b_2&pg={}'

# Encontrando o número de páginas
response = requests.get(url_base.format(1))
soup = BeautifulSoup(response.content, 'html.parser')
li_elements = soup.find_all('li', {'class': ['a-selected', 'a-normal', 'a-last']})
num_pages = len(li_elements)

# Iterando sobre cada página e coletando os dados dos livros
for page in range(1, num_pages+1):
    url = url_base.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for livro in soup.find_all('div', attrs={'class': 'p13n-gridRow _cDEzb_grid-row_3Cywl'}):
        for livro in soup.find_all('div',attrs={'class': 'a-column a-span12 a-text-center _cDEzb_grid-column_2hIsc'}):
            if livro.get('id') == 'gridItemRoot':
                livro_nome = [livro.find('div', attrs={'class': '_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y'}),livro.find('div', attrs={'class': '_cDEzb_p13n-sc-css-line-clamp-2_EWgCb'}),]
                if livro_nome is not None:
                    nome = livro_nome[1].text.strip() if livro_nome[1] is not None else livro_nome[0].text.strip()
                    link = livro.find('a', attrs={'class': 'a-link-normal'})['href']
                    livro_dict = {'nome': nome, 'link': link}
                    livros.append(livro_dict)

print(livros)
#print(len(livros))
```

# 1.2.2 - Principle Loop:

```python
# define o caminho do chromedriver
chromedriver_path = 'caminho/para/chromedriver'

driver = webdriver.Chrome(chromedriver_path)

# create an empty list to store the dictionaries -> https://www.amazon.com.br/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=2
df_livros = []

# loop through each link
for i, link in enumerate(livros):
    link_final = 'https://www.amazon.com.br' + link['link']
    print(f'Opening link {i+1} of {len(livros)}: {link_final}')
    
    # open the link in the browser
    driver.get(link_final)
    
    # capture the soup of the link opened
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    
    # create a dictionary to store the data for this book
    livro_dict = {}
    
    # get the name of the book
    try:
        nome_element = soup.find('span', attrs={'id': 'productTitle'})
        livro_dict['nome'] = nome_element.text.strip()
    except:
        livro_dict['nome'] = ''
    
    # get the price of the book
    try:
        preco_element = soup.find('span', attrs={'id': 'price'})
        livro_dict['preco'] = preco_element.text.strip()
    except:
        livro_dict['preco'] = ''
    
    # get the NPS of the book
    try:
        preco_element = soup.find('span', attrs={'id': 'acrCustomerReviewText'})
        livro_dict['avaliacao'] = preco_element.text.strip()
    except:
        livro_dict['avaliacao'] = ''
    
    try:
        autor_element = soup.find('span', {'class': 'author notFaded'}).find('a', {'class': 'a-link-normal'})
        autor_nome = autor_element.text.strip()
        livro_dict['autor'] = autor_nome
    except:
        livro_dict['autor'] = ''

    try:
        editora_element = soup.find('div', attrs={'id': 'rpi-attribute-book_details-publisher'}).find('div', attrs={'class': 'a-section a-spacing-none a-text-center rpi-attribute-value'})
        editora_nome = editora_element.text.strip()
        livro_dict['editora'] = editora_nome
    except:
        livro_dict['editora'] = ''
        
    try:
        pages_element = soup.find('div', attrs={'id': 'rpi-attribute-book_details-fiona_pages'}).find('div', attrs={'class': 'a-section a-spacing-none a-text-center rpi-attribute-value'})
        pages_nome = pages_element.text.strip()
        livro_dict['pages'] = pages_nome
    except:
        livro_dict['pages'] = ''
               
    try:
        clas_element = soup.find('div', attrs={'id': 'averageCustomerReviews'}).find('span', attrs={'class': 'a-declarative'}).find('span', attrs={'id': 'acrPopover'})
        clas_nome = clas_element.text.strip()
        livro_dict['NPS'] = clas_nome
    except:
        livro_dict['NPS'] = '0'
        
    try:
        publi_element = soup.find('div', attrs={'id': 'rpi-attribute-book_details-publication_date'}).find('div', attrs={'class': 'a-section a-spacing-none a-text-center rpi-attribute-value'})
        publi_nome = publi_element.text.strip()
        livro_dict['publication'] = publi_nome
    except:
        livro_dict['publication'] = ''
      
    # find the element that contains the stock availability
    try:
        estoque_element = soup.find('div', {'id': 'availability'})
        span_element = estoque_element.find('span', {'class': 'a-size-medium'})
        estoque = span_element.text.strip()
        livro_dict['estoque'] = estoque
    except:
        livro_dict['estoque'] = ''

    df_livros.append(livro_dict)

```

# 1.2.3 - Create, transform and prepare DataFrame:

```python
import numpy as np
from datetime import datetime

base = pd.DataFrame.from_dict(df_livros)
base['avaliacao'] = base['avaliacao'].apply(lambda x: x.split("avaliações de clientes")[0].strip()).astype(str).replace('',0).astype(str)
base['pages'] = base['pages'].apply(lambda x: x.split("páginas")[0].strip()).astype(str).astype(int)
base['NPS'] = base['NPS'].apply(lambda x: x.split("de 5 estrelas")[0].strip()).fillna(0).str.replace(',', '.').astype('float64')#.astype(int)
base['preco'] = base['preco'].str.replace('R\$', '').str.replace('.', '').str.replace(',', '.').str.strip()

# Substitui valores vazios por NaN
base['preco'] = base['preco'].replace('', 0).astype(float)
base['PricePerPages'] = ( base['preco'] / base['pages'])
base['FatPerAval'] = base['preco'].astype('float64') * base['avaliacao'].astype('float64')
base['avaliacao'] = base['avaliacao'].replace('', 0).astype(float).astype('float64')
base['data_base'] = datetime.now().strftime('%d-%m-%Y')
base['estoque'] = np.where(base['estoque'] == 'Em estoque', 'Em estoque','Sem estoque')

# Create classification
base['clas_price'] = np.where(base['preco'] <= 15, '0-15', np.where(base['preco'] <= 32, '15-32','+32'))
base['clas_Avaliacao'] = np.where(base['avaliacao'] <= 5.000, '-5.000', np.where(base['avaliacao'] <= 10.000, '5K - 10K','+10K'))

base[['dia_plubi', 'mes_plubi', 'ano_plubi']] = base['publication'].str.split(expand=True)
dic_plubi = {
    'janeiro':'01',
    'fevereiro':'02',
    'março':'03',
    'abril':'04',
    'maio':'05',
    'junho':'06',
    'julho':'07',
    'agosto':'08',
    'setembro':'09',
    'outubro':'10',
    'novembro':'11',
    'dezembro':'12'
}
base['month'] = base['mes_plubi'].apply(lambda x: dic_plubi[x])
base['day_publi'] = base['dia_plubi'].astype(str) + '-' +  base['month'].astype(str) + '-' + base['ano_plubi'].astype(str)
base = base[['nome', 'preco', 'avaliacao', 'autor', 'editora', 'pages', 'NPS', 'estoque', 'PricePerPages', 'FatPerAval', 'data_base','clas_price', 'clas_Avaliacao', 'day_publi']]

base['data_base'] = pd.to_datetime(base['data_base'], format='%d-%m-%Y')
base['day_publi'] = pd.to_datetime(base['day_publi'], format='%d-%m-%Y')
base['Var_Publi'] = base['data_base'] - base['day_publi']
base['Var_Publi'] = base['Var_Publi'].astype(str)
base['Var_Publi_Day'] = base['Var_Publi'].apply(lambda x: x.split("days")[0].strip()).astype(str).astype(int)
base['Var_Publi_Years'] = np.where(base['Var_Publi_Day'] <= 365, '1',base['Var_Publi_Day'] / 365)
base['Var_Publi_Years'] = base['Var_Publi_Years'].astype(float).round(0).astype(int)
base
```
